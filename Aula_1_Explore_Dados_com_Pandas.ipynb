**Traduza para mim todas as colunas do dataframes df para português brasileiro**.

```python
mapeamento_colunas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

df = df.rename(columns=mapeamento_colunas)
df.columns
```

```python
df["senioridade"].value_counts()
```

**Traduza para português as categorias da coluna "senioridade" do dataframe df.**

```python
mapeamento_senioridade = {
    'SE': 'Sênior',
    'MI': 'Intermediário',
    'EN': 'Júnior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(mapeamento_senioridade)

df["senioridade"].value_counts()
```

```python
df["contrato"].value_counts()
```

**Traduza para português as categorias da coluna "contrato" do dataframe df.**

```python
mapeamento_contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

df['contrato'] = df['contrato'].replace(mapeamento_contrato)

df['contrato'].value_counts()
```

```python
df["remoto"].value_counts()
```

**Traduza para português as categorias da coluna "remoto" do dataframe df.**

```python
mapeamento_remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

df['remoto'] = df['remoto'].replace(mapeamento_remoto)

df['remoto'].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

**Traduza para português as categorias da coluna "tamanho_empresa" do dataframe df.**

```python
mapeamento_tamanho_empresa = {
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(mapeamento_tamanho_empresa)

df['tamanho_empresa'].value_counts()
```
